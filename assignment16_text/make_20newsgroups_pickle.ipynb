# Convert newsgroups files to pickle
This will make reloading the data much faster!

```python
import pickle
import os
import sys

BASE_DIR = '/fs/scratch/PAS1495/physics6820/text/'
TEXT_DATA_DIR = os.path.join(BASE_DIR, '20news-18828')

# second, prepare text samples and their labels
print('Processing text dataset')

texts = []  # list of text samples
fnames = []  # list of filenames
labelnames = []  # list of labels
numTypes = 0
for name in sorted(os.listdir(TEXT_DATA_DIR)):
    numTypes += 1
    path = os.path.join(TEXT_DATA_DIR, name)
    print("   Processing: ",path)
    numFiles = 0
    if os.path.isdir(path):
        for fname in sorted(os.listdir(path)):
            if fname.isdigit():
                fpath = os.path.join(path, fname)
                args = {} if sys.version_info < (3,) else {'encoding': 'latin-1'}
                numFiles += 1
                with open(fpath, **args) as f:
                    t = f.read()
                    i = t.find('\n\n')  # skip header
                    if 0 < i:
                        t = t[i:]
                    texts.append(t)
                fnames.append(fname)
                labelnames.append(name)
        print("   ===> Files read:",numFiles)
    #if numTypes>2:
    #    break
print('Found %s texts.' % len(texts))
pickle.dump([fnames,labelnames,texts],open('20newsgroups_fnames_labelnames_texts.pkl', 'wb') )
```

```python
# Load library
from nltk.corpus import stopwords

# You will have to download the set of stop words the first time
import nltk
nltk.download('stopwords')
# Load stop words
stop_words = stopwords.words('english')
```

```python

# second, prepare text samples and their labels
print('Processing text dataset')

texts = []  # list of text samples
fnames = []  # list of filenames
labelnames = []  # list of labels
numTypes = 0
for name in sorted(os.listdir(TEXT_DATA_DIR)):
    numTypes += 1
    path = os.path.join(TEXT_DATA_DIR, name)
    print("   Processing: ",path)
    numFiles = 0
    if os.path.isdir(path):
        for fname in sorted(os.listdir(path)):
            if fname.isdigit():
                fpath = os.path.join(path, fname)
                args = {} if sys.version_info < (3,) else {'encoding': 'latin-1'}
                numFiles += 1
                with open(fpath, **args) as f:
                    t = f.read()
                    i = t.find('\n\n')  # skip header
                    if 0 < i:
                        t = t[i:]
                    words = t.lower().split()
                    newwords = []
                    #print(words)
                    for word in words:
                        if len(word)>0:
                            if word not in stop_words:
                                newwords.append(word)
                    new_t = ' '.join(newwords)
                    #print(new_t)
                    texts.append(new_t)
                fnames.append(fname)
                labelnames.append(name)
        print("   ===> Files read:",numFiles)
    #if numTypes>2:
    #    break
print('Found %s texts.' % len(texts))
pickle.dump([fnames,labelnames,texts],open('20newsgroups_fnames_labelnames_texts_nostopwords.pkl', 'wb') )
```
